```python
import numpy as np
import networkx as nx
```

```python
d = np.zeros((11, 11))
d
```

```python
d[1][2] = d[2][1] = d[3][0] = d[3][1] = d[4][1] = d[4][3] = d[4][5] = d[5][1] = d[5][4] = d[6][1] = d[6][4] = d[7][1] = d[7][4] = d[8][1] = d[8][4] = d[9][1] = d[9][4] = d[10][1] = d[10][4] = 1
```

```python
d
```

```python
G = nx.from_numpy_array(d, create_using=nx.DiGraph)
```

```python
G
```

```python
nx.draw_networkx(G)
```

```python
b = np.sum(d, axis = 1)
b[b == 0] = 1
stochastic = d/b[:,None]
stochastic = stochastic.T
```

```python
beta = 0.85
trap_resolve_mat = (1/len(d)) * np.ones((len(d), len(d)))
Google_Mat = beta * stochastic + (1 - beta) * trap_resolve_mat
Google_Mat
```

```python
def PageRank(mat, iter):
    vec = np.random.rand(len(d))
    for _ in range(iter):
        vec = np.matmul(mat, vec)
    return vec

page_rank = PageRank(Google_Mat, 50)
page_rank
```

```python
for idx, i in enumerate(page_rank):
    print(str(chr(idx + 65)) + ": " + str(i*100) + "%")
```
